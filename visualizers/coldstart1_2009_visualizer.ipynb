```python
import pandas as pd

records_path = "../records/coldstart1_2009_record_pre.tsv"

coldstart1_df = pd.read_csv(records_path, sep=",")

coldstart1_df
```

```python
coldstart1_df = coldstart1_df.loc[:, ['models', 'data_num', 'acc']]

coldstart1_df 
```

```python
dkt_df = coldstart1_df[coldstart1_df['models'] == 'dkt']
dkvmn_df = coldstart1_df[coldstart1_df['models'] == 'dkvmn']
sakt_df = coldstart1_df[coldstart1_df['models'] == 'sakt']
```

```python
dkt_df_idx = dkt_df[dkt_df['data_num'].str.contains('19917model_name')].index

dkt_df.drop(dkt_df_idx, inplace=True)

dkt_df.head()
```

```python
dkvmn_df_idx = dkvmn_df[dkvmn_df['data_num'].str.contains('19917model_name')].index

dkvmn_df.drop(dkvmn_df_idx, inplace=True)

dkvmn_df.head()
```

```python
#sakt_df

sakt_df_idx = sakt_df[sakt_df['data_num'].str.contains('19917model_name')].index

sakt_df.drop(sakt_df_idx, inplace=True)

sakt_df.tail()
```

```python
dkt_df = dkt_df.astype({'data_num':'int', 'acc': 'float'})
dkvmn_df = dkvmn_df.astype({'data_num':'int', 'acc': 'float'})
sakt_df = sakt_df.astype({'data_num':'int', 'acc': 'float'})
```

# 100 to 2000, 100개 단위

```python
dkt_df_100to2000 = dkt_df[dkt_df["data_num"] <= 2000]
dkvmn_df_100to2000 = dkvmn_df[dkvmn_df["data_num"] <= 2000]
sakt_df_100to2000= sakt_df[sakt_df["data_num"] <= 2000]
```

```python
dkt_df_100to2000
```

```python
#100~2000까지 100 단위로 비교
import matplotlib.pyplot as plt

plt.plot(dkt_df_100to2000["data_num"], dkt_df_100to2000['acc'], label = "dkt")
plt.plot(dkvmn_df_100to2000["data_num"], dkvmn_df_100to2000['acc'], label = "dkvmn")
plt.plot(sakt_df_100to2000["data_num"], sakt_df_100to2000['acc'], label = "sakt")
plt.legend(loc="lower right")
plt.show()
```

# 500 to 19500, 500 단위

```python
dkt_df_500to19500 = dkt_df[
    (dkt_df["data_num"] == 500) | (dkt_df["data_num"] == 1000) \
        | (dkt_df["data_num"] == 1500) | (dkt_df["data_num"] >= 2000)
    ]

dkvmn_df_500to19500 = dkvmn_df[
    (dkvmn_df["data_num"] == 500) | (dkvmn_df["data_num"] == 1000) \
        | (dkvmn_df["data_num"] == 1500) | (dkvmn_df["data_num"] >= 2000)
    ]

sakt_df_500to19500 = sakt_df[
    (sakt_df["data_num"] == 500) | (sakt_df["data_num"] == 1000) \
        | (sakt_df["data_num"] == 1500) | (sakt_df["data_num"] >= 2000)
    ]
```

```python
import matplotlib.pyplot as plt

plt.plot(dkt_df_500to19500["data_num"], dkt_df_500to19500['acc'], label = "dkt")
plt.plot(dkvmn_df_500to19500["data_num"], dkvmn_df_500to19500['acc'], label = "dkvmn")
plt.plot(sakt_df_500to19500["data_num"], sakt_df_500to19500['acc'], label = "sakt")
plt.legend(loc="lower right")
plt.show()
```
